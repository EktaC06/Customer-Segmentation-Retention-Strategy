```python
df = pd.read_csv('Telecom_Dataset_Clean.csv')
```

```python
service_flags = ['Multiple_Lines','Online_Security','Online_Backup','Device_Protection_Plan',
                  'Premium_Support','Streaming_TV','Streaming_Movies','Streaming_Music','Unlimited_Data']

df['Recency_Proxy'] = df['Tenure_in_Months']
df['Service_Count'] = df[service_flags].apply(lambda row: (row == 'Yes').sum(), axis=1)
df['Frequency_Proxy'] = df['Number_of_Referrals'] + df['Service_Count']
df['Monetary_Proxy'] = df['Total_Revenue']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

rfm_features = ['Recency_Proxy','Frequency_Proxy','Monetary_Proxy']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[rfm_features])

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
rfm_features = ['Recency_Proxy','Frequency_Proxy','Monetary_Proxy']

profile = df.groupby('Segment')[rfm_features + ['Customer_Status']].agg({
    'Recency_Proxy':'mean',
    'Frequency_Proxy':'mean',
    'Monetary_Proxy':'mean',
    'Customer_Status': lambda x: (x=='Churned').mean()*100
}).round(2)
profile.columns = ['Avg_Tenure_Months','Avg_Frequency','Avg_Revenue','Churn_Rate_%']
profile['Count'] = df['Segment'].value_counts().sort_index()

print(profile)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []
K_range = range(1, 11)
for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```

```python
segment_labels = {
    0: 'Steady Long-Tenure (At Risk)',   # your Seg0: tenure 28.9, churn 31%
    1: 'New & Engaged (At Risk)',        # your Seg1: tenure 9.5, churn 32%
    2: 'High-Value Champions',           # your Seg2: revenue 7510, churn 14.7%
    3: 'New & Disengaged'                # your Seg3: freq 5.3 (lowest), churn 27.9%
}
df['Segment_Label'] = df['Segment'].map(segment_labels)
df.to_csv('Telecom_Final_Segmented.csv', index=False)
```
